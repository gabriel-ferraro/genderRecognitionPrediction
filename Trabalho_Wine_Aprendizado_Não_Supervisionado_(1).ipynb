# Análise dos Componentes Principais (PCA)

**Extração de Características:** Encontra os relacionamentos dos melhores atributos e cria novos atributos.

É um algoritmo de aprendizagem não supervisionada.

Aplica-se em dados linearmente separáveis.

O PCA utiliza a variância explicada: Quanto cada atributo contríbui para a variação total no dataSet. Dessa forma, é posível 'rankear' os atributos mais relvantes se baseando no percentual de variância, quanto maior mais relevante.

```python
from sklearn.decomposition import PCA
```

O PCA é especialmente útil para dimensionar dados; valores em diferentes medidas podem ser dimensionados para uma dimensão desejada, desconsiderando a necessidade do conhecimento de grandezas e medidas por parte do desenvolvedor da análise.

```python
pca = PCA(n_components=2) #setando a dimensionalidade para duas dimensões
```

```python
pca_transform = pca.fit_transform(df)
```

```python
pca_transform.shape
```

```python
pca_transform
```

```python
pca.explained_variance_ratio_
```

A soma das variâncias individuais de cada elemento do dataSet resultam em uma variância explicada de 99.4% Valor ótimo para criar um modelo de predição de alta acurácia.

```python
print("Taxa de variância explicada dos dados em duas dimensões: %.2f%%" % (pca.explained_variance_ratio_.sum() * 100) )
```

```python
import plotly.express as px
```

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    pca_transform,
    labels=labels,
    dimensions=range(2), # numero de componentes do pca (PCA(n_components=2))
    color=df["qualidade"]
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

# **K-Means com dois atributos**

K-means é um algoritmo de clusters (agrupa dados em subgrupos). Para isso é necessário escalonar os dados do dataSet.

**Escalonamento**

```python
from sklearn.preprocessing import StandardScaler
escala = StandardScaler()
df_esc = escala.fit_transform(df)
```

**K-Means**

```python
dados = df.iloc[:, [5,6]].values
dados
```

Para usar o K-means é necessário fornecer a quantidade de clusters k. Gerar a quantidade otimizada de clusters é possível com um método como o elbow method. O elbow method demonstra o decaimento da variação da soma dos quadrados em seus respectivos clusters em função da quantidade de clusters. O local onde o decaimento é mais acentuado forma algo semelhante ao cotovelo de um braço, o valor nesse ponto é a quantidade de clusters 'ideal'.

**Elbow Method**

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 5, max_iter = 300)
    kmeans.fit(dados)

    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Com base no elbow method foi decidido que 3 seria um bom número de clusters

```python
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS');
```

**Agrupamento**

O algoritmo gera o posicionamento dos centróides, os dados mais próximos são atribuídos a esses centróides, depois em repetidas iterações os centróides são movidos de acordo com a média aritmética da distância entre esses dados e cada centróide. Ao alcançar a quantidade de iterações estabelecida ou um estado em que os dados não mudam mais de centróide, é determinado a qual cluster cada dado pertence.

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 5, max_iter = 300)
kmeans1 = kmeans.fit(dados)
```

```python
# Coordenadas dos Centróides
centroides = kmeans1.cluster_centers_
centroides
```

```python
# Classificação dos dados
classificacao = kmeans1.labels_
classificacao
```

# **Agrupamento Hierárquico: todos atributos**


A predição para um dataSet com alta dispersão é mais eficiente com o uso de PCA, no caso do uso de agrupamento hierárquico pode se gerar uma árvore com distância elevada em razão da quantidade de agrupamentos.

```python
plt.figure(figsize=(10, 10))
dendrograma = dendrogram(linkage(df_esc, method = 'ward')) #ward (dados dispersos)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hier = AgglomerativeClustering(n_clusters=12, affinity='euclidean', linkage = 'ward')
classificacao = hier.fit_predict(df_esc)
```

```python
classificacao
```

**Agrupamento**

```python
agrupamento = pd.DataFrame(classificacao, columns = ['grupo'])
agrupamento
```
